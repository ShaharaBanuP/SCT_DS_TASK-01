```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("C:API_SP.POP.TOTL_DS2_en_csv_v2_71683.csv",skiprows=4)
df
```

```python
df.info
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=[col for col in df.columns if 'Unnamed' in col])
```

```python
df=df.drop(columns=['2024'])
```

```python
#choose a year to visualise
year='2020'
```

```python
df['2020'].dtypes
```

```python
df[year] = pd.to_numeric(df[year], errors='coerce')
```

```python
df_clean=df.dropna(subset=[year])
```

### Histogram for Population Distribution in 2020

```python
plt.figure(figsize=(10,6))
sns.histplot(df_clean[year],bins=30,kde=True)
plt.title(f'Population Distribution in {year}')
plt.xlabel('Number of Countries')
plt.ylabel('Population')
plt.show()
```

### Bar Chart: Top 10 Most Populous Contries

```python
top10=df_clean.sort_values(by=year,ascending=False).head(10)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=top10[year],y=top10['Country Name'],palette='viridis')
plt.title(f'10 Top Most populous Contries in {year}')
plt.xlabel('Population')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```
